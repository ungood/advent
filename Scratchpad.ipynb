```python
def reverse_string(s):
  # Create a list of characters in the string
  chars = list(s)

  # Reverse the list of characters
  chars.reverse()

  # Create a new string by joining every other character
  # from the reversed list
  return ''.join([chars[i] for i in range(0, len(chars), 2)])


reverse_string('4b756970')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def prob(cache, turn):
    if turn == 0:
        return (0.6, 0.5)
    
    prev_turn = cache[turn-1]
        
    a = prev_turn[0] * (1.0 - prev_turn[1])
    b = prev_turn[1] * (1.0 - prev_turn[0])
    
    return (a, b)

turns = np.arange(0, 100)
cache = []
for turn in turns:
    cache.append(prob(cache, turn))
    
a = [foo[0] for foo in cache]
b = [foo[1] for foo in cache]

plt.plot(turns, a, c='red')
plt.plot(turns, b)
```
