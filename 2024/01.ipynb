```python
import aocd
from utilities import *

puzzle = aocd.get_puzzle(year=2024, day=1)

display_puzzle(puzzle)
```

## Part A

```python
def parse(data_in):
    data_line_split = data_in.split('\n')
    data_split = []
    data_left = []
    data_right = []
    
    for line in data_line_split:
        values = line.split('   ')
        data_left.append(int(values[0]))
        data_right.append(int(values[1]))
        
    return data_left, data_right


assert parse(puzzle.examples[0].input_data) == ([3, 4, 2, 1, 3, 3], [4, 3, 5, 3, 9, 3])


def align(data_left, data_right):
    sorted_left = sorted(data_left)
    sorted_right = sorted(data_right)
    return sorted_left, sorted_right

assert align(*parse(puzzle.examples[0].input_data)) == ([1, 2, 3, 3, 3, 4], [3, 3, 3, 4, 5, 9])

def diff(data_left, data_right):
    differences = []
    
    for left, right in zip(data_left, data_right):
        differences.append(abs(left - right))
    
    return differences

assert diff(*align(*parse(puzzle.examples[0].input_data))) == [2, 1, 0, 1, 2, 5]

def add(differences):
    total = 0
    
    for difference in differences:
        total += difference
    return total

assert add(diff(*align(*parse(puzzle.examples[0].input_data)))) == int(puzzle.examples[0].answer_a)

add(diff(*align(*parse(puzzle.input_data))))
```

## Part 2

```python
def find_number(number, data_right):
    num_count = 0
    for value in data_right:
        if value == number:
            num_count += 1
    return num_count

assert find_number(57964, parse(puzzle.input_data)[1]) == 11

def addify(data_left, data_right):
    score = 0
    
    for left in data_left:
        occurences = find_number(left, data_right)
        
        if occurences != 0:
            score += left * occurences
    return score

assert addify(*parse(puzzle.examples[0].input_data)) == 31

addify(*parse(puzzle.input_data))
```
