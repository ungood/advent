```python
import aocd
from utilities import *

puzzle = aocd.get_puzzle(year=2024, day=2)

display_puzzle(puzzle)
```

## Part A

```python
def parse(unusual_data):
    '''Turns unusable unusual data into usable unusual data'''
    reports = []
    unusual_data = unusual_data.split('\n')
    
    for report in unusual_data:
        parsed_report = []
        values = report.split(' ')
        
        for value in values:
            parsed_report.append(int(value))
            
        reports.append(parsed_report)

    return reports

assert parse(puzzle.examples[0].input_data) == [[7, 6, 4, 2, 1], [1, 2, 7, 8, 9], [9, 7, 6, 2, 1], [1, 3, 2, 4, 5], [8, 6, 4, 4, 1], [1, 3, 6, 7, 9]]

def issafe(report):
    '''Turns usable unusual report into usable unusual number'''
    direction = None
    for i in range(1, len(report)):
        difference = report[i] - report[i - 1]
        if difference <= 3 and difference > 0 and direction != "down":
            direction = "up"
        elif difference >= -3 and difference < 0 and direction != "up":
            direction = "down"
        else:
            return 0
    return 1
    

assert issafe([7, 6, 4, 2, 1]) == 1
assert issafe([1, 3, 6, 7, 9]) == 1
assert issafe([9, 7, 6, 2, 1]) == 0
assert issafe([1, 1, 1, 1, 1]) == 0

def aresafe(unusual_data):
    '''Turns usable unusual data into usable unusual reports into usable unusual number into usable unusual answer!'''
    total_safe = 0
    usable_unusual_data = parse(unusual_data)
    for report in usable_unusual_data:
        total_safe += issafe(report)
    return total_safe

aresafe(puzzle.input_data)
```

## Part 2

```python
def couldbesafe(report):
    '''We told it it was wrong, but it convenced us to let it try again'''
    if issafe(report) == 1:
        return 1
    else:
        for i in range(len(report)):
            better_now = []
            for j in range(len(report)):
                if j != i:
                    better_now.append(report[j])
            if issafe(better_now) == 1:
                return 1
    return 0

assert couldbesafe([1, 3, 2, 4, 5]) == 1

def aresafe_fineyoucantryagain(unusual_data):
    '''Turns usable unusual data into usable unusual reports into usable unusual number into usable unusual answer!'''
    total_safe = 0
    usable_unusual_data = parse(unusual_data)
    for report in usable_unusual_data:
        total_safe += couldbesafe(report)
    return total_safe

aresafe_fineyoucantryagain(puzzle.input_data)
```
